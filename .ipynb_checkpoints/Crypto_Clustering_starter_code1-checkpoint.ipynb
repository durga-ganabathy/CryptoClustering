```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
# import os
# os.environ["OMP_NUM_THREADS"] = "1"
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
df_market_data.columns
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data[["price_change_percentage_24h" , "price_change_percentage_7d", 
                                                                       "price_change_percentage_14d", "price_change_percentage_30d",
                                                                       "price_change_percentage_60d", "price_change_percentage_200d", 
                                                                       "price_change_percentage_1y"]])

# Diplay the first three rows of the scaled data
market_data_scaled[0:3]
```

```python
df_market_data_scaled = pd.DataFrame(
    market_data_scaled,
    columns=["price_change_percentage_24h" , "price_change_percentage_7d", 
            "price_change_percentage_14d", "price_change_percentage_30d",
            "price_change_percentage_60d", "price_change_percentage_200d", 
            "price_change_percentage_1y"])

# df_market_data_scaled

# Copy the crypto names from the original data
df_market_data_scaled["coin_id"] = df_market_data.index

# Set the coinid column as index
df_market_data_scaled = df_market_data_scaled.set_index("coin_id")

# Display sample data
df_market_data_scaled.head()


# CORRECT CODE BELOW
# df_market_data_scaled = pd.DataFrame(
#     market_data_scaled,
#     columns=["price_change_percentage_24h" , "price_change_percentage_7d", 
#             "price_change_percentage_14d", "price_change_percentage_30d",
#             "price_change_percentage_60d", "price_change_percentage_200d", 
#             "price_change_percentage_1y"]
# )

# # Get the list of coin IDs
# coins = list(df_market_data.index)

# # Copy the crypto names from the original data
# df_market_data_scaled["coin_id"] = coins

# # Set the coin_id column as index
# df_market_data_scaled.set_index("coin_id", inplace=True)

# # Display the columns of the DataFrame
# # print(df_market_data_scaled.columns)
# df_market_data_scaled.head()
```

```python
df_market_data_scaled.columns
```

# Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python

# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_market_data_scaled)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Review the DataFrame
df_elbow.head(10)

```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot_orgdata = df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Plot using Original data", 
    xticks=k
)
elbow_plot_orgdata
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** From the elbow curve, the optimal value for `k` is 4. Therefore, the number of clusters would be 4.
 
 To be noted: The inertia value for the k=4 is 79.022435.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
cluster_group = model.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
cluster_group
```

```python
# Create a copy of the DataFrame
market_data_scaled_predictions_df = df_market_data_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
market_data_scaled_predictions_df['cluster group'] = cluster_group
# market_data_scaled_predictions_df.set_index('coin_id', inplace=True)
# Display sample data
market_data_scaled_predictions_df
```

```python
market_data_scaled_predictions_df.columns
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
scatter_org_data = market_data_scaled_predictions_df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="cluster group",
    hover_cols="coin_id",
    title = "Scatter Plot using Original data"
)
scatter_org_data 
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
marketdata_pca = pca.fit_transform(market_data_scaled_predictions_df)
# View the first five rows of the DataFrame. 
marketdata_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** About 89% (0.36941241+ 0.29169375+ 0.22893878) of the total variance is condensed into the 3 PCA variables.
About 88.67% (0.37269822+ 0.32489961+ 0.18917649) of the total variance is condensed into the 3 PCA variables.  About 89% (0.34871677+ 0.31363391+ 0.22627118) of the total variance is condensed into the 3 PCA variables.

```python
# Create a new DataFrame with the PCA data.
df_market_data_pca = pd.DataFrame(marketdata_pca, columns=["PCA1", "PCA2", "PCA3"])

# Review the PCA DataFrame
df_market_data_pca.head()

# Copy the crypto names from the original data
df_market_data_pca["coin_id"] = df_market_data.index

# Set the coinid column as index
df_market_data_pca = df_market_data_pca.set_index("coin_id")

# Display sample data
df_market_data_pca.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    model = KMeans(n_clusters=i, random_state=1)
    model.fit(df_market_data_pca)
    inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {
    "k": k,
    "inertia": inertia
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca = pd.DataFrame(elbow_data_pca)
df_elbow_pca.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot_pca = df_elbow_pca.hvplot.line(x="k", y="inertia", title="Elbow plot Using PCA Data", xticks=k)
elbow_plot_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:**  4 is the best value for `k` using the PCA data.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** The `k` value is the same in both the cases(original and PCA). However, inertia from the original data is high (Datasets with low concentration of elements) when compared to the inertia from the PCA data(low inertia - Datasets with a high concentration of elements). Hence the elements in the scatter plot from the original data is spread out. Whereas in the scatter plot from PCA data, the elements are tightly grouped together.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the PCA data
model.fit(df_market_data_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
crypto_clusters = model.predict(df_market_data_pca)
# Print the resulting array of cluster values.
crypto_clusters
```

```python
# Create a copy of the DataFrame with the PCA data
df_market_pca_predictions = df_market_data_pca.copy()

# Add a new column to the DataFrame with the predicted clusters
df_market_pca_predictions["CryptoCluster"] = crypto_clusters

# Display sample data
df_market_pca_predictions.head(10)


```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
scatter_plot_pca = df_market_pca_predictions.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="CryptoCluster",
    hover_cols= "coin_id",
    title = "Scatter Plot using PCA data"
)
scatter_plot_pca
```

#### Answer the following question: 

  * **Question:** After visually analyzing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

  * **Answer:** In both the cases, whether using original data or using PCA data the `k` value determined by the elbow method is 4. So, when it comes to clustering both the datasets exhibit similar underlying structures. However the inertia with k=4 for the original data is 79 and for the PCA data is 44. Low inertia value reflects tighter and more compact clusters indicating better clustering performance. By using few features, the dimensionality of the data is reduced, making it easier for the K-Means to identify meaningful clusters. Hence using PCA data for dimensionality reduction and feature extraction provides effective clustering results.
